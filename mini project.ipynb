```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
days=['mon','tue','wed','thur','fri','sat','sun']
temperature=np.array([30, 32, 34, 33, 35, 36, 37])
humidity=np.array([45, 50, 55, 53, 52, 48, 47])
wind_speed=np.array([10,12,14,11,10,9,8])
```

```python
print('Average temperature:',np.mean(temperature))
print('Average humidity:',np.mean(humidity))
print('Average wind speed:',np.mean(wind_speed))
print('max temp:',np.max(temperature),'min temp:',np.min(temperature))
print('max humidity:',np.max(humidity),'min humidity:',np.min(humidity))
print('max wind speed:',np.max(wind_speed),'min wind speed:',np.min(wind_speed))
print('Standard deviation of temp:',np.std(temperature))
s=np.where(temperature==np.max(temperature))
v=int(s[0][0])
print('On',days[v],'the temperature was maximum')
c=np.where(humidity==np.max(humidity))
g=int(c[0][0])
print('On',days[g],'the humidity was maximum')
norm_temp=(temperature-temperature.mean())/temperature.std()
print("after normalizing data",norm_temp)
```

```python
plt.plot(days,temperature, 'g*--')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.grid()
plt.show()
```

```python
plt.plot(days, temperature,'g*-',label='temperature')
plt.plot(days, humidity,'r^-',label='humidity')
plt.plot(days, wind_speed,label='wind speed')
plt.xlabel('days')
plt.ylabel('values')
plt.grid()
plt.title('temp ,hum and wind speed analysis' )
plt.legend()
plt.show()
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.grid(axis='y')
plt.title('temperature analysis')
plt.bar(days,temperature)
plt.show()
```

```python
plt.pie(temperature,labels=days , autopct='%0.1f%%',explode=[0,0,0,0,0,0,0.4])
plt.axis='equal'
plt.show()
```

```python
plt.hist(temperature, bins=[30,32,34,36,38],rwidth=0.95, label='temperature')
plt.xlabel('temperature')
plt.ylabel('distribution')
plt.title('Distr of temperature')
plt.grid()
plt.show()
```
